```python
#Exercise 1: Basic Usage of SciPy
import scipy
print("SciPy version:", scipy.__version__)
```

```python
#Exercise 2: Descriptive Statistics
data = [12, 15, 13, 12, 18, 20, 22, 21]
import numpy as np
from scipy import stats

mean = stats.tmean(data)
median = np.median(data)
variance = stats.tvar(data)
std_dev = stats.tstd(data)

mean, median, variance, std_dev
```

```python
#Exercise 3: Understanding Distributions
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(10, 90, 1000)
y = norm.pdf(x, loc=50, scale=10)

plt.plot(x, y)
plt.title("Normal Distribution (Mean=50, Std=10)")
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
#Exercise 4: T-Test Application
from scipy.stats import ttest_ind
import numpy as np

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = ttest_ind(data1, data2)

t_stat, p_value
```

```python
#Exercise 5: Linear Regression Analysis
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]

from scipy.stats import linregress

result = linregress(house_sizes, house_prices)

slope = result.slope
intercept = result.intercept

slope, intercept


predicted_price = slope * 90 + intercept
predicted_price
```

```python
#Exercise 6: Understanding ANOVA
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
from scipy.stats import f_oneway

f_value, p_value = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

f_value, p_value
```

```python
# Exercise 7: Probability Distributions
from scipy.stats import binom

probability = binom.pmf(5, n=10, p=0.5)
probability
```

```python
#Exercise 8: Correlation Coefficients
import pandas as pd

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})
from scipy.stats import pearsonr, spearmanr

pearson_corr, _ = pearsonr(data['age'], data['income'])
spearman_corr, _ = spearmanr(data['age'], data['income'])

pearson_corr, spearman_corr
```
