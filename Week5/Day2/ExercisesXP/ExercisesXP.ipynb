```python
#Exercise 1 : Matrix Operations
import numpy as np


A = np.array([[2, 1, 3],
              [4, 1, 6],
              [7, 8, 9]])

det_A = np.linalg.det(A)

inv_A = np.linalg.inv(A)

print("Matrix A:\n", A)
print("\nDeterminant:", det_A)
print("\nInverse:\n", inv_A)
```

```python
# Exercise 2 : Statistical Analysis
data = np.random.rand(50)

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python
#Exercise 4 : Data Manipulation with NumPy and Pandas
import pandas as pd

df = pd.DataFrame(np.random.randint(1, 100, (5, 4)),
                  columns=['A', 'B', 'C', 'D'])


selected = df[df['A'] > 50]

total_sum = df.sum()
average = df.mean()

print("DataFrame:\n", df)
print("\nA > 50:\n", selected)
print("\nSum:\n", total_sum)
print("\nAverage:\n", average)
```

```python
#Exercise 6 : Basic Hypothesis Testing
productivity_before = np.random.normal(50, 10, 30)
productivity_after = productivity_before + np.random.normal(5, 3, 30)

mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)

print("Mean before:", mean_before)
print("Mean after:", mean_after)

if mean_after > mean_before:
    print("Training appears to improve productivity.")
else:
    print("No improvement observed.")
```

```python
#Exercise 8 : Time Series Data Manipulation
dates = np.arange('2023-01', '2024-01', dtype='datetime64[M]')
values = np.random.rand(12)

time_series = pd.Series(values, index=dates)

print("Jan–Mar:\n", time_series['2023-01':'2023-03'])
print("\nApr–Jun:\n", time_series['2023-04':'2023-06'])
print("\nJul–Sep:\n", time_series['2023-07':'2023-09'])
print("\nOct–Dec:\n", time_series['2023-10':'2023-12'])
```
