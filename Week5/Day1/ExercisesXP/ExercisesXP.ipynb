```python
import numpy as np
```

```python
#Exercice 1:Array Creation and Manipulation
arr1 = np.arange(10)
print(arr1)
```

```python
#Exercise 2:Type Conversion and Array Operations
lst = [3.14, 2.17, 0, 1, 2]
arr2 = np.array(lst, dtype=int)
print(arr2)
```

```python
#Exercise 3:  Working with Multi-Dimensional Arrays
arr3 = np.arange(1, 10).reshape(3, 3)
print(arr3)
```

```python
#Exercise 5 : Indexing Arrays
array = np.array([[21,22,23,22,22],
                  [20,21,22,23,24],
                  [21,22,23,22,22]])

second_row = array[1]
second_row

```
