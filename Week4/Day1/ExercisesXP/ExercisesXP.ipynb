```python
#Exercise 2:Creating a Line Plot for Temperature Variation
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.figure()
plt.plot(days, temperatures, marker='o')
plt.xlabel("Day")
plt.ylabel("Temperature (°F)")
plt.title("Weekly Temperature Variation")
plt.show()
```

```python
#Exercise 3: Visualizing Monthly Sales with a Bar Chart
import matplotlib.pyplot as plt


months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [5000, 5500, 6200, 7000, 7500]


plt.figure()
plt.bar(months, sales)
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales Data")
plt.show()
```

```python
#Exercise 4 : Data Visualisation
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data.csv")

df.head()
total_quantity = df["quantity"].sum()
print("Total quantity of products sold:", total_quantity)

revenue_by_category = df.groupby("category")["revenue"].sum()

highest_category = revenue_by_category.idxmax()
highest_revenue = revenue_by_category.max()

print("Category with highest revenue:", highest_category)
print("Revenue generated:", highest_revenue)
average_revenue = df["revenue"].mean()
print("Average revenue per sale:", average_revenue)

df["date"] = pd.to_datetime(df["date"])
df["quarter"] = df["date"].dt.to_period("Q")
quarterly_revenue = df.groupby("quarter")["revenue"].sum()
quarterly_revenue
plt.figure()
quarterly_revenue.plot(kind="bar")
plt.xlabel("Quarter")
plt.ylabel("Total Revenue")
plt.title("Total Revenue Generated per Quarter")
plt.show()
```
