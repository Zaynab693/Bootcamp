```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import random
import os
from scipy.ndimage import rotate
from PIL import Image
```

```python

IMAGE_DIR = "DailyChallenge"

# check folder exists
if not os.path.exists(IMAGE_DIR):
    print(f"Folder '{IMAGE_DIR}' not found!")
    print("Creating folder...")
    os.makedirs(IMAGE_DIR)
else:
    print(f"Folder '{IMAGE_DIR}' found!")

image_files = [
    f for f in os.listdir(IMAGE_DIR)
    if f.lower().endswith(('.jpg', '.png', '.jpeg'))
]

print("Images found:", len(image_files))

list_img = []
list_names = []

plt.figure(figsize=(20,20))
for i in range(min(5, len(image_files))):
    file = random.choice(image_files)
    img_path = os.path.join(IMAGE_DIR, file)
    img = mpimg.imread(img_path)
    list_img.append(img)
    list_names.append(file)

for i, img in enumerate(list_img):
    ax = plt.subplot(3,5,i+1)
    ax.set_title(list_names[i])
    plt.imshow(img)
    plt.axis("off")
plt.show()
```

```python
#Rotation de 30°
from scipy.ndimage import rotate

def rotate_image_30_degrees(image):
    return rotate(image, 30, reshape=False)
plt.show()
```

```python
#Flip horizontal + vertical
import numpy as np

def flip_image(image):
    return np.flipud(np.fliplr(image))

plt.figure(figsize=(10,10))

for i, img in enumerate(list_img):
    flipped = flip_image(img)

    plt.subplot(2,3,i+1)
    plt.title("Flip H + V")
    plt.imshow(flipped)
    plt.axis("off")

plt.show()
```

```python
#Zoom ×1.2
from PIL import Image

def zoom_image(image, scale=1.2):
    h, w = image.shape[:2]
    new_h, new_w = int(h*scale), int(w*scale)

    img_pil = Image.fromarray((image*255).astype(np.uint8))
    zoomed = img_pil.resize((new_w, new_h))

    left = (new_w - w)//2
    top = (new_h - h)//2
    right = left + w
    bottom = top + h

    return np.array(zoomed.crop((left, top, right, bottom)))
```

```python
#Visualisation : original vs augmented (side by side)
plt.figure(figsize=(15,10))

for i, img in enumerate(list_img):
    plt.subplot(5,2,2*i+1)
    plt.title("Original")
    plt.imshow(img)
    plt.axis("off")

    plt.subplot(5,2,2*i+2)
    plt.title("Rotation 30°")
    plt.imshow(rotate_image_30_degrees(img))
    plt.axis("off")

plt.show()
```
