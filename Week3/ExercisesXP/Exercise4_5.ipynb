```python

#Exercice 5
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris_df = pd.read_csv(r"C:\Users\asus\Desktop\GenAI\Week3\ExercicesXP\Iris_Dataset\Iris.csv")
iris_df.head()
```

```python
# Mean
mean_val = iris_df['SepalLengthCm'].mean()
# Median
median_val = iris_df['SepalLengthCm'].median()
# Mode
mode_val = iris_df['SepalLengthCm'].mode()[0]  # mode() returns a Series

print(f"SepalLengthCm - Mean: {mean_val}, Median: {median_val}, Mode: {mode_val}")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(iris_df['SepalLengthCm'], bins=10, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
