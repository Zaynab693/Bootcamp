```python
def classify_columns(df):
    print(f"Dataset: {df.name}" if hasattr(df, 'name') else "Dataset:")
    for col in df.columns:
        if df[col].dtype == 'object':
            print(f"{col} → Qualitative (categorical)")
        else:
            print(f"{col} → Quantitative (numerical)")
    print("\n")


sleep_df.name = "Sleep Dataset"
mental_df.name = "Mental Health Dataset"
credit_clean_df.name = "Credit Card Clean Dataset"
credit_crx_df.name = "Credit Card CRX Dataset"


classify_columns(sleep_df)
classify_columns(mental_df)
classify_columns(credit_clean_df)
classify_columns(credit_crx_df)
```
